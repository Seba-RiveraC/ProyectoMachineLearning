```python
# 1️⃣ Inicializar Kedro y cargar el catálogo
from kedro.framework.session import KedroSession

session = KedroSession.create("..")
context = session.load_context()
catalog = context.catalog
print("✅ Kedro listo")

# 2️⃣ Cargar la tabla combinada (merge de PIB per cápita + esperanza de vida)
df = catalog.load("merged_data")

print(f"✅ Datos cargados correctamente: {df.shape[0]} filas y {df.shape[1]} columnas")
df.head()
```

```python
# Ver estructura y tipos de datos
print("Tipos de datos:")
df.info()

# Resumen estadístico
print("\nEstadísticas descriptivas:")
display(df.describe(include="all"))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="whitegrid")

plt.figure(figsize=(14, 6))

# Distribución de la esperanza de vida
plt.subplot(1, 2, 1)
sns.histplot(df["period_life_expectancy_at_birth"], kde=True, color="skyblue", bins=30)
plt.title("Distribución de la Esperanza de Vida")

# Distribución del PIB per cápita
plt.subplot(1, 2, 2)
sns.histplot(df["gdp_per_capita"], kde=True, color="salmon", bins=30)
plt.title("Distribución del PIB per cápita")

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(
    data=df,
    x="gdp_per_capita",
    y="period_life_expectancy_at_birth",
    hue="region",
    alpha=0.7
)
plt.title("PIB per cápita vs Esperanza de Vida por Región")
plt.xlabel("PIB per cápita (US$ actuales)")
plt.ylabel("Esperanza de Vida al Nacer (años)")
plt.xscale("log")
plt.legend(title="Región", bbox_to_anchor=(1.05, 1), loc="upper left")
plt.show()
```

```python
numeric_cols = df.select_dtypes(include=["float64", "int64"])
corr = numeric_cols.corr()

plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Matriz de Correlación de Variables Numéricas")
plt.show()

# Mostrar correlación PIB per cápita vs Esperanza de vida
corr_target = corr["period_life_expectancy_at_birth"].sort_values(ascending=False)
print("\n📊 Correlaciones con la esperanza de vida:")
display(corr_target)
```
